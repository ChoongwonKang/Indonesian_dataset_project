```python
#문장 분리 후 'Sentence' column에 적용
def onebyone(test):
    test = re.sub(r'&#\d+;', "", test) #remove &#38;(문장 부호 유형)
    test = re.sub(r'#\d+;', "", test) #remove #8216;, #8217; etc(문장 부호 유형)
    test = re.sub(r'# \d+;', "", test)
    test = re.sub(r'#\d+ ;', "", test)
    test = re.sub(r'# \d+ ;', "", test)
    test = re.sub(r"'#\d+;[^']+'", "", test)
    test = re.sub(r"ig_mid=[A-Z0-9\-]+", "", test) # #38; 뒤에 오는 ig_mid= 형태의 문장 제거(문장 부호 유형)
    test = re.sub(r'\[ \d+ \] : .*',"",test) #remove [ 1 ] : [ 2 ] : [ 3 ] etc
    test = re.sub(r"Written by:.*","",test) #Written by: 이후 모든 문장 제거(출처&링크 유형)
    test = re.sub(r"Written by :.*","",test)
    test = re.sub(r"Written by.*","",test)
    test = re.sub(r"written by:.*","",test)
    test = re.sub(r"written by :.*","",test)
    test = re.sub(r"written by .*","",test)
    test = re.sub(r"—- Written by :.*","",test)
    test = re.sub(r"—Written by :.*","",test)
    test = re.sub(r"#main #main #main.*","",test) # #main 이후 모든 문장 제거
    test = re.sub(r"About these ads.*","",test) #About these ads. 이후 모든 문장 제거(광고 유형)
    test = re.sub(r"Allkpop.*","",test) #Allkpop 이후 모든 문장 제거(제거 시 가끔 이메일 주소도 같이 삭제)
    test = re.sub(r"allkpop.*","",test)
    test = re.sub(r"Indotrans.*","",test) #allkpop 바로 뒤에 나오는 indotrans 이후 모든 문장 제거
    test = re.sub(r"indotrans.*","",test)
    test = re.sub(r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.(com|net|kr)\b","",test) #remove email(.com, .net, .kr)
    test = re.sub(r"\([^@]+@[^\)]+\)","",test) #(@e-mail) 형태 문장 제거
    test = re.sub(r"\(@[^)]+\)","",test) #(@단어) 형태 문장 제거(SNS 계정 포함)
    test = re.sub(r"\( @[^)]+\)","",test) 
    test = re.sub(r"\(#[^)]+\)","",test) #(#단어) 형태 문장 제거
    test = re.sub(r"@\w+\b","",test) #@포함 바로 뒤에 오는 단어 제거
    test = re.sub(r"#\w+\b","",test) # #포함 바로 뒤에 오는 단어 제거
    test = re.sub(r"utm_medium.*","",test) #utm_medium 이후 모든 문장 제거 (Soompi 관련 text)
    test = re.sub(r"utm_campaign.*","",test)
    test = re.sub(r"Photo Credit.*","",test) #Photo Credit/credit 이후 모든 문장 재거
    test = re.sub(r"Photo credit.*","",test)
    test = re.sub(r"Article credit.*","",test)
    test = re.sub(r"Lihat videonya.*","",test) #광고문 뒷문장 모두 제거
    test = re.sub(r"lihat videonya.*","",test) 
    test = re.sub(r"^.*\d+\s*points\s*\d+.*$","",test) #'숫자 points 숫자' 형태의 문장 제거(고유명사 나열 유형)
    test = re.sub(r"^.*\d+\s*Points\s*\d+.*$","",test) 
    test = re.sub(r"^.*\d+\s*points\s*\.*$","",test)
    test = re.sub(r"^.*\d+\s*poin.*$","",test) # '숫자 poin' 형태의 문장 제거(고유명사 나열 유형)
    test = re.sub(r"^.*\d+\s*Poin.*$","",test)
    test = re.sub(r"^.*berikut\sini\.\s*$","",test) #텍스트 마지막에 존재하는 berikut ini. 가 포함된 문장 제거(인니어 광고문 유형)
    test = re.sub(r"^.*bawah ini.*$","",test) #di bawah ini가 포함된 문장 제거(인니어 광고문 유형)
    test = re.sub(r"^.*\.\.\.$","",test) # ...으로 끝나는 문장 제거(문장 잘림 유형)
    test = re.sub(r"\[.*?\]","",test) # 대괄호 및 단어 제거
    test = re.sub(r"www\.[a-zA-Z0-9-]+\.com","",test) # www.com 형태의 문장 제거
    test = re.sub(r"\b\w+\.com\b","",test) # .com 형태의 문장 제거
    test = re.sub(r"\([^)]*persen[^)]*\)","",test) #(persen...text) 형태의 문장 제거
    test = re.sub(r"\([^)]*Entertainment[^)]*\)","",test) #(Entertainment...text) 형태의 문장 제거
    test = re.sub(r".+\s*\.\.\.$","",test) #마지막이 ...로 끝나는 문장 모두 제거

    #하나씩 제거하는 정규표현식들은 맨 마지막에 적용

    test = re.sub(r'gt;_lt;', '',test) #remove ;gt and ;lt
    test = re.sub(r'gt;',"",test)
    test = re.sub(r'lt;',"",test)
    test = re.sub(r'_lt;',"",test)
    test = re.sub(r"Advertisements","",test) #remove english
    test = re.sub(r"Instagrammable","",test)
    test = re.sub(r"instagrammable","",test)
    test = re.sub(r"LOL","",test)
    test = re.sub(r"LoL","",test)
    test = re.sub(r"gantiin","",test) #remove wrong spelling
    test = re.sub(r"content]","",test)
    test = re.sub(r"@ star1","",test) #@star1 이라는 잡지회사 제거
    test = re.sub(r"@star1","",test)
    test = re.sub(r"♥","",test) #특수문자 제거
    test = re.sub(r"▲","",test)
    test = re.sub(r"=","",test)
    test = re.sub(r"#","",test) 
    test = re.sub(r"@","",test) 
    test = re.sub(r"\[","",test) 
    test = re.sub(r"]","",test) 
    test = re.sub(r"~","",test) 
    test = re.sub(r"\^","",test) 
    test = re.sub(r"→","",test) 
     
    return test
```

```python
#정규표현식을 적용한 'Sentence' column에 있는 텍스트를 토크나이징
def tokenizing_sent(df):
    new_rows = []
    max_length = 5
    for index, row in tqdm(df.iterrows(), total=df.shape[0], desc="Processing rows"):
        article_sent = row['Sentence']
        stanza_doc = nlp(article_sent)
        tokens = [token.text for sent in stanza_doc.sentences for token in sent.tokens]
        new_row = row.copy()
        new_row['Token']= tokens
        new_row['Word_Count'] = len(tokens)
        new_rows.append(new_row)
    
    new_df = pd.DataFrame(new_rows)
    new_df['Tokenized_Sentence']= new_df['Token'].apply(' '.join)

    new_df.drop_duplicates(subset='Sentence', inplace=True)
    
    new_df.reset_index(drop=True, inplace=True)
    new_df.rename(columns={'Source_File' : 'Filename', 'Headline':'Title', 'Keyword':'Pub_Subj', 'Date': 'Pub_Date'}, inplace=True)
    
    new_df = new_df[['Doc_ID', 'Filename', 'Title', 'Pub_Type', 'Pub_Subj', 'Pub_Date', 'Col_Date', 'Sen_ID', 'Word_Count', 'Text', 'Sentence','Tokenized_Sentence' , 'Token']]
    new_df = new_df[new_df['Token'].apply(lambda x: len(x) > max_length)]
    return new_df
```

```python
def cleansing(text):
    # "MMM DD, YYYY" 형식의 날짜와 ")"로 끝나는 부분을 찾아 제거하는 정규표현식
    pattern_1 = r'\w{3}\s\d{2},\s\d{4}\(.*?\)'
    pattern_2 = r'Loading Link to the original story\. Notes.*?sole discretion\.'
    pattern_3 = r'\b[A-Z][a-z]{2} \d{2}, \d{4} \(.*?by Newstex\)'
    
    # 정규표현식을 사용하여 패턴을 제거
    cleaned_text = re.sub(pattern_3, '', text, flags=re.DOTALL)
    
    # cleaned_text = re.sub(pattern_1, '', text)

    # 왼쪽 공백 제거
    cleaned_text = cleaned_text.lstrip()

    if text.endswith('sole discretion.'):
        cleaned_text= cleaned_text[:-1284]
    

    # 날짜 형식 제거 (예: Apr 08, 2013)
    cleaned_text = re.sub(r'\b[A-Z][a-z]{2} \d{2}, \d{4}\b', '', cleaned_text)
    # http:// 또는 https:// 형태 제거
    cleaned_text= re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', cleaned_text)
    # 대괄호 안 숫자 제거
    cleaned_text = re.sub(r'\[\d+\]', '', cleaned_text)
    # <--> 형태 제거
    cleaned_text = re.sub(r'<.*?>', '', cleaned_text)
    

    # 정규표현식 패턴을 사용하여 해당 패턴 제거
    pic_pattern = re.compile(r'pic\.twitter\.com\/[A-Za-z0-9]+')
    cleaned_text = re.sub(pic_pattern, "", cleaned_text)


    # 한글 제거
    text_without_korean = re.sub('[가-힣]+', '', cleaned_text)
    ko_pattern = re.compile(r'[ㄱ-ㅎㅏ-ㅣ]+')
    text_without_korean = re.sub(ko_pattern, '', text_without_korean)
    
    
    # 링크 제거 (http:// 또는 https://로 시작하는 링크)
    text_without_links = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', text_without_korean)
    if (text_without_links.startswith("(")) & (")" in text_without_links):
        end_index = text_without_links.index(')')
        text_without_links = text_without_links[end_index+1:]
    if text_without_links.startswith("(KoreanIndo: Delivered by Newstex)"):
        text_without_links = text_without_links[len("(KoreanIndo: Delivered by Newstex)"):]
    
    
    pattern = re.compile(r'(source|Source)')
    match = pattern.search(text_without_links)

    
    if match:
        text_without_links = re.split(r'(source|Source).*', text_without_links)[0].strip()
        
    else:
        pass
    text_without_links = text_without_links.replace("Loading", "")
    
    
    text_without_links = text_without_links.replace("[embedded content]", "")
    text_without_links = re.sub(r'\[ \d+\]:', '', text_without_links)
    return text_without_links
```
